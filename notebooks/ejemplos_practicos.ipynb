```python
def categorialize_price(price):
    if price < 3:
        return 'Low'
    elif price < 5:
        return 'Medium'
    else:
        return 'High'
    
df['Price_Category'] = df['UnitPrice'].apply(categorialize_price)
print('DataFrame con nueva columna Price_Category: \n')
print(df[['UnitPrice', 'Price_Category']].head(10))
```

```python
# Crear DataFrames de ejemplo
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value2': [4, 5, 6]
})

# Realizar un merge interno
inner_merged = pd.merge(df1, df2, on='key', how='inner')
print("Unión Interna:\n", inner_merged)
```

```python
# Realizar un merge externo
outer_merged = pd.merge(df1, df2, on='key', how='outer')
print("Unión Externa:\n", outer_merged)
```

```python
# Realizar un merge izquierda
left_merged = pd.merge(df1, df2, on='key', how='left')
print("Unión Izquierda:\n", left_merged)
```

```python
# Realizar un merge derecha
right_merged = pd.merge(df1, df2, on='key', how='right')
print("Unión Derecha:\n", right_merged)
```

```python
# Crear DataFrames de ejemplo
df3 = pd.DataFrame({
        'A': ['A0', 'A1', 'A2'],
        'B': ['B0', 'B1', 'B2']
    })
    
df4 = pd.DataFrame({
        'A': ['A3', 'A4', 'A5'],
        'B': ['B3', 'B4', 'B5']
    })
    
# Concatenar verticalmente
vertical_concat = pd.concat([df3, df4])
print("Concatenación Vertical:\n", vertical_concat)
```

```python
plt.Figure(figsize=(8,6))

categories = ['producto A', 'producto B', 'producto C', 'producto D', 'producto E']
sales = [20, 35, 30, 35, 27]

max_ventas = max(sales)

colores = []

for v in sales:
    if v == max_ventas:
        colores.append('red')  # color destacado
    else:
        colores.append('gray')  # los demás

#creacion del gráfico de barras
plt.barh(categories, sales, color=colores, label='Ventas Mensuales')

plt.title('Ventas Mensuales por Producto')
plt.xlabel('Ventas')
plt.ylabel('Productos')
plt.legend()

plt.show()
```

```python
from matplotlib.dates import DateFormatter

dates = pd.date_range(start='2025-01-01', periods=100, freq='D')

values = np.random.rand(100).cumsum()

data = pd.DataFrame({'Date': dates, 'Value': values})
```

```python
#grafico de lineas
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(data['Date'], data['Value'], color = 'purple')
ax.set_xlabel('Fecha')
ax.set_ylabel('Valor')
ax.set_title('Evolución del Valor a lo largo del tiempo')
plt.show()
```

```python
dates = pd.date_range(start='2025-01-01', periods=12, freq='ME')
sales = np.random.randint(1000, 5000, size=12)

sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})

plt.plot(sales_data['Date'], sales_data['Sales'], marker='o', linestyle='-', color='orange', label='Ventas Mensuales')

plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

plt.xticks(rotation=45)

plt.title('Ventas Mensuales')
plt.xlabel('Fecha')
plt.ylabel('Ventas')

plt.legend()

plt.tight_layout()

plt.show()
```
