```python
df ['Total_Price'] = df['Quantity'] * df['UnitPrice']
print('DataFrame con nueva columna Total_Price: \n')
print(df.head())
```

```python
df['high_Value'] = df['Total_Price'] > 16
print('DataFrame con nueva columna high_Value: \n')
print(df['high_Value'].head(10))
```

```python
df['Discounted_Price'] = df['UnitPrice'].apply(lambda x: x * 0.9)
print('DataFrame con nueva columna Discounted_Price: \n')
print(df.head(5))
```

```python
def categorialize_price(price):
    if price < 3:
        return 'Low'
    elif price < 5:
        return 'Medium'
    else:
        return 'High'
    
df['Price_Category'] = df['UnitPrice'].apply(categorialize_price)
print('DataFrame con nueva columna Price_Category: \n')
print(df[['UnitPrice', 'Price_Category']].head(10))
```

```python
country_counts = df['Country'].value_counts()
print('Número de transacciones por país: \n')
print(country_counts)
```

```python
country_groups = df.groupby('Country')['Quantity'].sum()
print('Cantidad total por país: \n')
print(country_groups)
```

```python
country_stats = df.groupby('Country')['UnitPrice'].agg(['mean', 'min', 'max', 'sum'])
print('Estadísticas de precio unitario por país: \n')
print(country_stats)
```

```python
country_stock_group = df.groupby(['Country', 'StockCode'])['Quantity'].sum()
print('Cantidad total por país y código de producto: \n')
print(country_stock_group)
```

```python
def total_revenue(group):
    return (group['Quantity'] * group['UnitPrice']).sum()

revenue_by_country = df.groupby('Country').apply(total_revenue)
print('Ingresos totales por país: \n')
print(revenue_by_country)
```

```python
ventas_pais = df.groupby('Country')['Total_Price'].sum()
ventas_pais = ventas_pais.reset_index()
ventas_pais = ventas_pais.sort_values(by='Total_Price', ascending=False)

print("Top 3 mayores ventas:\n", ventas_pais.head(3))
print("\n Top 3 menores ventas:\n", ventas_pais.tail(3))
```

```python
#filtrar ventas en el Reino Unido

file_path = '../data/online_retail.csv'
sales_data = pd.read_csv(file_path)

uk_sales = sales_data[sales_data['Country'] == 'United Kingdom']
print('Ventas en el Reino Unido: \n')
print(uk_sales)
```

```python
high_quantity_sales = df[df['Quantity'] > 10]
print('Ventas con cantidad mayor a 10: \n')
print(high_quantity_sales)
```

```python
uk_high_quantity_sales = sales_data[(sales_data['Country'] == 'United Kingdom') & (sales_data['Quantity'] > 50)]
print('Ventas en el Reino Unido con cantidad mayor a 50: \n')
print(uk_high_quantity_sales)
```

```python
# Convertir a fecha
sales_data['InvoiceDate'] = pd.to_datetime(sales_data['InvoiceDate'])

# Filtrar ventas del 2011
sales_2011 = sales_data[sales_data['InvoiceDate'].dt.year == 2011]

print('Ventas en 2011:\n')
print(sales_2011)
```

```python
sales_diciembre_2010 = sales_data[(sales_data['InvoiceDate'].dt.year == 2010) & (sales_data['InvoiceDate'].dt.month == 12)]
print('Ventas en diciembre de 2010: \n')
print(sales_diciembre_2010)
```
